```python
def normal_prompt(question):

    prompt = f"""
    Answer the following question:

    {question}
    """

    return prompt
```

```python
def cot_prompt(question):

    prompt = f"""
    You are a logical reasoning engine.

    Follow these steps:

    Step 1: Understand the problem.

    Step 2: Break the problem into smaller parts.

    Step 3: Solve each part.

    Step 4: Verify your reasoning.

    Step 5: Give the final answer.

    Question:
    {question}
    """

    return prompt
```

```python
def ask_gemini(prompt):

    response = model.generate_content(prompt)

    return response.text
```
